```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('smartphone_sales-1.csv')
```

```python
import statsmodels.formula.api as smf
```

```python
linear_reg = smf.ols("sales ~ Prices + display_size + primary_camera + storage + battery", data=df)
```

```python
result = linear_reg.fit()
result.summary()
```

```python
from linearmodels import IV2SLS # Instrumental Variables two stage least squares
iv_regression = IV2SLS.from_formula("sales ~ [Prices ~ CNYExchangeRateIndex] + display_size + primary_camera + storage + battery", data=df)
results_iv = iv_regression.fit()

# Because we used a intrumental variable CNYExchangeRateIndex, we can now interpret the coefficient of Prices as causal effect of Prices on sales
```
