```python
import numpy as np

def estimate_black_jack_prob(n=1000000):
    deck = [2,3,4,5,6,7,8,9,10,10,10,10,11]
    deck = deck*4
    succ = 0
    for i in range(n):
        hand = np.random.choice(deck, 2, replace=False)
        if sum(hand) == 21:
            succ += 1
    return succ/n
```

```python
estimate_black_jack_prob()
```
